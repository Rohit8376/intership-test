#### Q2. Now consider "filteredCountry.csv" as the input file. For each group of "SKU" find 2 minimum prices and store this result in "lowestPrice.csv" inside a folder named "output"

```python
data = pd.read_csv("output/filteredCountry.csv")
```

```python
data.sort_values(by=['SKU']).head(50)
```

```python
df  = data.iloc[:,[0,5]]
df
```

```python
count = 0
value12 = {}
list1 = []
for i in df.index:
    if not df['SKU'][i] in value12.keys():
        value12[df['SKU'][i]] =[df['PRICE'][i]]
    else:
        value12[df['SKU'][i]].append(df['PRICE'][i])
    list1.append(df['PRICE'][i])
```

```python
for i in value12:   
    for j in range(len(value12[i])):
        value12[i][j] = value12[i][j].replace("?","")
        value12[i][j] = value12[i][j].replace("$","")
        value12[i][j] = float(value12[i][j].replace(",",""))
        
```

```python
correct_data = {}
for i in value12:
    if len(value12[i]) > 1:
        correct_data[i] = value12[i]
        correct_data[i].sort()

lowestPrice = ""
for i in correct_data:
    lowestPrice = lowestPrice+str(i)+","+str(correct_data[i][0]) + ","+str(correct_data[i][1])+","

  
```

```python
lowestPrice
        
        
        
```

```python
lowestPrice = lowestPrice.split(',');lowestPrice.remove("")

```

```python
import numpy as np
lowestPrice = pd.DataFrame(np.array(lowestPrice).reshape(int(len(lowestPrice)/3),3), columns = ["SKU","FIRST_MINIMUM_PRICE","SECOND_MINIMUM_PRICE"] )
```
